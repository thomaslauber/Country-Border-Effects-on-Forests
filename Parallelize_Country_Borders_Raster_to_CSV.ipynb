```python
import ee as ee
ee.Initialize()
```

```python
composite = ee.Image("users/devinrouth/ETH_Composites/CrowtherLab_Composite_30ArcSec")
countryNeighbors = ee.Image("users/devinrouth/ETH_Country_Borders/Country_Neighbors_100km")
countriesImage = ee.Image("users/devinrouth/Countries_of_the_World_30ArcSec")
customImages = ee.Image("users/devinrouth/ETH_Country_Borders/20191210_CountryBordersCustomLayers_BordersOnly_EqualArea");

unboundedGeo = ee.Geometry.Polygon([-180, 88, 0, 88, 180, 88, 180, -88, 0, -88, -180, -88], None, False);

finalImage = ee.Image.cat(customImages,composite.select(['Pixel_Lat','Pixel_Long']),countriesImage,countryNeighbors).updateMask(customImages.select('Forest_Total_Potential').add(100));
print(finalImage.bandNames().getInfo())

vectors = ee.FeatureCollection('users/devinrouth/ETH_Country_Borders/20191217_Tiles_for_Country_Borders_Export');

def rasterToTabular(f):
    tabularData = ee.FeatureCollection(finalImage.reduceRegion(reducer=ee.Reducer.toCollection(finalImage.bandNames()),
                                                               geometry=f.geometry(),
                                                               maxPixels=1e13).values()).flatten();
    return tabularData
```

```python
listOfFeatures = ee.Dictionary(vectors.aggregate_histogram('label')).keys().getInfo();
testListOfFeatures = listOfFeatures
len(testListOfFeatures)
```

```python
for label in testListOfFeatures:
    dataToExport = vectors.filterMetadata('label','equals',int(label)).map(rasterToTabular).flatten()
    
    task = ee.batch.Export.table.toCloudStorage(
        collection = dataToExport,
        description = 'CountryBorders_Feature_'+label+'_',
        bucket = 'theforgottenrealms27182818'
        )
    task.start()
    print('CountryBorders_Feature_'+label+'_')
```
