```r
tic()
# Apply the function to every raw dataset
pb = txtProgressBar(min=0, max=length(fileList), initial=0, style=3)
emptyList = list()
for (i in fileList){
    setTxtProgressBar(pb,which(fileList==i))
    tryCatch({
    rawData = read.csv(paste('Raw_Data/',i,sep=''))
    cleanedData = cleanData(rawData)
    write.csv(cleanedData,paste('Processed_Data/',i,sep=''),row.names=FALSE)
    emptyList = c(emptyList,i)
    }, error=function(e){})
}
cat('\n')
toc()

```
